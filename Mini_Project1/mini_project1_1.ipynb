```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression, SGDClassifier


x,y = make_classification(n_samples=1000, n_features=2, n_redundant=0, n_classes=2, n_clusters_per_class=1, class_sep=1.5, random_state=93)
print(x.shape,y.shape)

plt.scatter( x[:,0], x[:,1], c=y)
```

Q1_2:

Logistic Regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split


df = pd.read_csv("data_banknote_authentication.txt")
r1 = list(df.columns)
print(r1)
df.columns=['variance of Wavelet Transformed image','skewness of Wavelet Transformed image','curtosis of Wavelet Transformed image','entropy of image','class']
df.loc[len(df)] = r1
df
```

Q2_2:

```python
from sklearn.model_selection import train_test_split

train,test = train_test_split(df, test_size=0.3, shuffle=True, random_state=93)
print(train,"\n",test)
```

Q2_3:

```python
x_train = np.array([train['variance of Wavelet Transformed image'].values, train['skewness of Wavelet Transformed image'].values,
     train['curtosis of Wavelet Transformed image'].values, train['entropy of image'].values])
print(x_train.T.shape)

y_train = np.array(train[['class']].values)
print(y_train.shape)

x_test = np.array([test['variance of Wavelet Transformed image'].values, test['skewness of Wavelet Transformed image'].values,
     test['curtosis of Wavelet Transformed image'].values, test['entropy of image'].values])
print(x_test.T.shape)

y_test = np.array(test[['class']].values)
print(y_test.shape)


def sigmoid(x):
  x = x.astype(float)
  n = 1/(1 + np.exp(-x))
  return n

def Logistic_regression(w, p):
  p = p.astype(float)
  y_hat = sigmoid(p @ w)
  return y_hat

def BCE(y, y_hat):
  Loss = -(np.mean(y*np.log(y_hat)+(1-y)*np.log(1-y_hat)))
  return Loss

def GD(x, y, y_hat):
  G = (x @ (y_hat-y)) / len(y)
  return G

def optimizer(w ,eta, G):
  w = w - eta*G
  return w

def accuracy(y, y_hat):
  acc = np.sum (y == np.round(y_hat)) / len(y)
  return acc

epochs = 500
w = np.random.randn(4,1)
error = []

print(f"w1={w}")
print(w.shape)

for i in range(epochs):

  y_hat = Logistic_regression(w, x_train.T)

  L = BCE(y_train,y_hat)
  error.append(L)

  gradient = GD(x_train, y_train, y_hat)
  eta = 0.05
  w = optimizer(w, eta, gradient)

  if (i+1) % 100 == 0:
    print(f'Epoch={i}')
    print(f"Loss = {L}")


acc = accuracy(y_train, y_hat)
print(f"train accuracy = {acc}")

y_hat = Logistic_regression(w, x_test.T)
acc2 = accuracy(y_test, y_hat)
print(f"test accuracy = {acc2}")

```

```python
import matplotlib.pyplot as plt

plt.plot(error)
```

```python
def Normalization(x, x_max, x_min):
  norm = (float(x)- x_min) / (x_max - x_min)
  return norm

Max = []
Min = []

for j in range(4):
  max = 0
  min = 0

  for i in df[df.columns[j]]:
    if float(i) > max:
      max = i
    if float(i) < min:
      min = i

  Max.append(max)
  Min.append(min)

print(Max)
print(Min)

df_n = df.copy()
for j in range(4):
  df1 = []
  for i in range(len(df)):
    n = Normalization(df_n[df_n.columns[j]][i], Max[j], Min[j])
    df1.append(n)
  df_n[df_n.columns[j]] = df1

df_n
```
