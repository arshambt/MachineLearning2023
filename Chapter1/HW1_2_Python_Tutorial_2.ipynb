**Problem 10:** Given a Pandas DataFrame containing information about students, add a new column 'Grade' based on the following criteria:
* If 'Score' >= 90, Grade is 'A'
* If 'Score' >= 80, Grade is 'B'
* If 'Score' >= 70, Grade is 'C'
* If 'Score' < 70, Grade is 'D'

````python
import pandas as pd


data = {
    "Student" : ["Steven", "Melissa", "Aron", "Beth", "Bryan"],
    "Field" : ["Bio", "Math", "Math","Bio","Math"],
    "Score": [83, 92, 77, 88, 65]
}

df = pd.DataFrame(data)

grade = []

for i in range(len(df["Student"])):
  if df["Score"][i] >= 90:
    grade.append("A")
  elif df["Score"][i] >= 80:
    grade.append("B")
  elif df["Score"][i] >= 70:
    grade.append("C")
  elif df["Score"][i] < 70:
    grade.append("D")

df["Grade"] = grade

print(df)
````
